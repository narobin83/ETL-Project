```python
import pandas as pd
from sqlalchemy import create_engine
file_name = "oscars.csv"
df = pd.read_csv("oscars.csv")
df
```

```python
#drop colum 2 
new_df = df.drop(columns=['year_ceremony','ceremony', 'Race', 'gender', 'name'])
new_df
```

```python
new_df = new_df.rename(columns={"film":"title","year_film":"year","Category":"category"})
new_df
```

```python
# lowercase Category column 
new_df['category'] = new_df['category'].str.capitalize()

new_df['category'].unique()
```

```python
new_df = new_df.dropna(subset=["title"],how="any")
```

```python
new_df = new_df.drop_duplicates(subset=["title"],keep="first")
```

```python
# Database Setup
username = 'postgres'
password = 'postgres'

engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/movies_ETL')
rds_connection_string ="postgres:postgres@localhost:5432/movies_ETL"
```

```python
new_df.to_sql(name="oscars", con=engine, if_exists= "append", index=False)
```
