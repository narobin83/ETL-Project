```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "rotten_tomatoes_top_movies.csv"
rotten_movies_df = pd.read_csv(csv_file)
rotten_movies_df.head()
```

```python
duplicate_title = rotten_movies_df.loc[rotten_movies_df.duplicated(subset=['title',]),'title'].unique()
duplicate_title
```

```python
clean_df = rotten_movies_df[rotten_movies_df['title'].isin(duplicate_title)==False]
clean_df
```

```python
new_rotten_movies_df = clean_df[['title', 'critic_score', 'people_score', 'total_reviews', 'rating']].copy()
new_rotten_movies_df
```
