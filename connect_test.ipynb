```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "rotten_tomatoes_top_movies.csv"
rotten_movies_df = pd.read_csv(csv_file)
rotten_movies_df.head()
```

```python
new_rotten_movies_df = rotten_movies_df[['title', 'critic_score', 'people_score', 'total_reviews', 'rating']].copy()
new_rotten_movies_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/movies_ETL"
```

```python
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_rotten_movies_df.to_sql(name="rotten_tomatoes_data",con=engine, if_exists="append", index=False)
```
